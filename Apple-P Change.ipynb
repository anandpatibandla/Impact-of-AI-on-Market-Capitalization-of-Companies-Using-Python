```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from pandas.compat import PY310

# Fetch Apple stock data
apple = yf.Ticker("AAPL")
apple_data = apple.history(period="6mo")

# Convert the article date to a pandas datetime object
# Specify the timezone (e.g., 'America/New_York')
if PY310:
    article_date = pd.Timestamp("2024-04-11", tz="America/New_York")
else:
    article_date = pd.Timestamp("2024-04-11", tz=pd.Timezone("America/New_York"))

# Calculate the stock price difference before and after the article date
before_article = apple_data.loc[:article_date]
after_article = apple_data.loc[article_date:]

# Price change in dollars for the stock
price_change_dollars = after_article["Close"].iloc[1] - before_article["Close"].iloc[-1]

# Calculate the percentage change in stock price
price_change = (after_article["Close"].iloc[1] - before_article["Close"].iloc[-1]) / before_article["Close"].iloc[-1] * 100

# Print the percentage change
print(f"The stock price of Apple changed by {price_change:.2f}% after the article was published.")
print(price_change_dollars)

# Plot the stock data
plt.figure(figsize=(12, 6))
plt.plot(apple_data.index, apple_data["Close"])
plt.axvline(article_date, color="r", linestyle="--", label="Article Date")
plt.title("Apple Stock Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```
